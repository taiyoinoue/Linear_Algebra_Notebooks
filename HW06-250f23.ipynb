## Linear Transformations

**Definition:** A **linear transformation** $L$ between vector spaces $V$ and $W$ is a function from $V$ to $W$ satisfying the following two properties:
1. For all $\vec{v_1}$ and $\vec{v_2} \in V$, $L(\vec{v_1} + \vec{v_2}) = L(\vec{v_1}) + L(\vec{v_2})$
2. For all $\vec{v} \in V$ and $r \in \mathbb{R}$, $L(r \cdot \vec{v}) = r\cdot L(\vec{v})$.

**Example** (The most important example) Suppose $A$ is an $m \times n$ matrix.  Then we can define a linear transformation $L_A$ from $\mathbb{R}^n$ to $\mathbb{R}^m$ in the following way:
$$L_A(\vec{v}) = A \vec{v}$$
It is easy to check that $L_A$ is additive.  This follows from the fact that matrix multiplication distributes over matrix addition.  It is also easy to check that $L_A$ repsects scalar multiplication.  
In terms of notation, it is common to denote the linear transformation $L_A$ with just $A$.  Although $A$ is really just a matrix, it is so often thought of as a linear transformation that there is often no need for new notation to indicate this.

**Example** (The identity map) For any vector space $V$, define the **identity map** from $V$ to itself to be the function: $$\mathrm{Id}_V: V \to V$$
given by $\mathrm{Id}_V(\vec{v}) = \vec{v}$ for all $\vec{v} \in V$.  This is a linear transformation.

**Example** Consider $P_n = \{ \text{polynomials of degree at most $n$ in the variable $x$ with real coefficients} \}$.  This is a vector space of dimension $n + 1$. A basis is $\{ x^n, x^{n-1}, \dots, x, 1\}$.  Define the function:
$$\rm{Eval_2}: P_n \to \mathbb{R}$$
by
$$\rm{Eval_2}(p(x)) = p(2)$$
This is a linear transformation since: 
- $\rm{Eval_2}(p(x) + q(x)) = p(2) + q(2) = \rm{Eval_2}(p(x)) + \rm{Eval_2}(q(x))$
- $\rm{Eval_2}(r\cdot p(x)) = r \cdot p(2) = r\cdot \rm{Eval_2}(p(x))$

Therefore, this function respects both addition and scalar multiplication and so is a linear transformation.

**Example** A familiar example of a linear transformation is:
$$\dfrac{d}{dx}: P_n \to P_{n - 1}$$
given by:
$$\dfrac{d}{dx}(p(x)) = \dfrac{dp}{dx}$$
That is, this function outputs the deriviative of the polynomial.  Note that the derivative decreases the degree of a polynomial by 1, and so it is appropirate to make $P_{n -1}$ the codomain of this function.

### Other properties of linear transformations
- The composition of linear transformations is again a linear transformation.  More precisely, if $L: V \to W$ and $M: W \to U$ are linear transformations, then:
$$M \circ L: V \to U$$ 
is also a linear transformation.
- Linear transformations are determined by their action on a basis.  More precisely, if $\mathcal{B}$ is a basis of $V$, and $L: V \to W$ is a linear transformation, then you can determine what $L$ does to any vector in $V$ if you know what it does to the vectors in $\mathcal{B}$. 

### Representation of linear transformations with respect to bases

Recall that if we have a basis $\mathcal{B}$ of a vector space $V$, then we can define a function:
$$\mathrm{Rep}_{\mathcal{B}}: V \to \mathbb{R}^n$$ 
where $n$ is the dimension of $V$.  Recall that $\mathrm{Rep}_{\mathcal{B}}(\vec{v})$ is defined to be the column vector whose entries, when used as the coefficients of a linear combination of the elements of $\mathcal{B}$ (which can only be done in exactly one way since $\mathcal{B}$ is a basis), is equal to $\vec{v}$.  

In fact, this map is a linear transformation!  You proved this in the last homework set where you showed that $\mathrm{Rep}_{\mathcal{B}}$ respects addition and scalar multiplication.  

We might think of this map as a way of *representing* vectors in a vector space $V$ with respect to a basis $\mathcal{B}$.  But it turns out we can also represent linear transformations which map from $V$ to $W$ with respect to bases $\mathcal{B}$ on $V$ and $\mathcal{D}$ on $W$.  Rather than obtaining column vectors, we obtain matrices.

Given a linear transformation $L: V \to W$, and bases $\mathcal{B} \subset V$ and $\mathcal{D} \subset W$, we define the **representation of $L$ with respect to $\mathcal{B}$ and $\mathcal{D}$** to be the matrix:

$$\mathrm{Rep}_{\mathcal{B}\mathcal{D}}(L) = [ \mathrm{Rep}_\mathcal{D}(L(\vec{v_1})) \, \mid \, \mathrm{Rep}_\mathcal{D}(L(\vec{v_2})) \,  \mid \, \dots \, \mid \, \mathrm{Rep}_\mathcal{D}(L(\vec{v_n})) ]$$

where $\mathcal{B} = \{ \vec{v_1}, \dots, \vec{v_n} \} $. Notice that this matrix is $m \times n$ where $m = \dim W$ and $n = \dim V$, so we can view it as a linear transformation:
$$\mathrm{Rep}_{\mathcal{B}\mathcal{D}}(L): \mathbb{R}^n \to \mathbb{R}^m$$
in the usual way.

What is beautiful about this is the way that $\mathrm{Rep}_{\mathcal{B}\mathcal{D}}(L)$ interacts with both the maps $\mathrm{Rep}_{\mathcal{B}}$ and $\mathrm{Rep}_{\mathcal{D}}$.  This is described by the following theorem:

**Theorem:** For all $\vec{v} \in V$: 
$$\mathrm{Rep}_{\mathcal{B}\mathcal{D}}(L) \cdot \mathrm{Rep}_{\mathcal{B}}(\vec{v}) = \mathrm{Rep}_{\mathcal{D}}(L(\vec{v}))$$

We can summarize this theorem using the following **commutative square for the representation of $L$ with respect to $\mathcal{B}$ and $\mathcal{D}$**:



We call this a *commutative* diagram to emphasize that it doesn't matter how you move around the diagram (assuming you're following the directions of the arrows of course) -- you will always end up with the same result.  That the diagram is commutative is the content of the theorem above.

### Representation of a composition of linear transformations

**Theorem** Suppose $L: V \to W$ and $M: W \to U$ are linear transformations.  Then $M \circ L : V \to U$ is a linear transformation.  If $\mathcal{B}$, $\mathcal{C}$ and $\mathcal{D}$ are bases of $V, W$ and $U$ respectively, then 
$$\mathrm{Rep}_{\mathcal{B}\mathcal{D}}(M \circ L) = \mathrm{Rep}_{\mathcal{C}\mathcal{D}}(M)\cdot \mathrm{Rep}_{\mathcal{B}\mathcal{C}}(L)$$

- How can we check if a collection of vectors in $\mathbb{R}^n$ is a basis?  Recall that we have a theorem tailor made for just such a situation: 

    **Theorem:** A collection of vectors is a basis of $\mathbb{R}^n$ if and only if the matrix $A$ formed by using the vectors as columns is non-singular.
    
    Therefore, we see that having a function like `is_nonsingular` could be useful here!  Here, I'll write this function for you.  Run the cell below to bring it into memory for your use later on.

````python
import numpy as np

def is_nonsingular(array):
    if np.shape(array)[0] != np.shape(array)[1]:
        return False
    else:
        if np.linalg.det(array) != 0:
            return True
        else:
            return False
````

Here is another function you might find useful.  I call it `cat_col()`.  Why call it this? Because 'cat' is short for *concatenate*, which is an oft-used word in computer programming when joining two objects together.  So, for example, if we concatenate the strings `'hello'` and `'world'` we would get the string `'helloworld'`.  So `cat_col()` concatenates the column vectors from the list that we will use as input to the function:

````python
def cat_col(list):
    k = len(list)
    A = list[0]
    for i in range(1,k):
        A = np.append(A, list[i], 1)
    return A 
````

## Change of Basis

Suppose that $L: V \to W$ is a linear transformation, and that $\mathcal{B}, \mathcal{B}'$ are two bases for $V$ and $\mathcal{D}$ and $\mathcal{D}'$ are two bases for $W$. An important question to ask is, "How does the matrix $\mathrm{Rep}_{\mathcal{B}\mathcal{D}}(L)$ change if we represent the linear transformation $L$ with respect to bases $\mathcal{B}'$ and $\mathcal{D}'$?  Is there a formula for $\mathrm{Rep}_{\mathcal{B}'\mathcal{D}'}(L)$ in terms of $\mathrm{Rep}_{\mathcal{B}\mathcal{D}}(L)$?

We call this the question of **change of basis**, hopefully for obvious reasons.

To answer this question, consider the identity maps of $V$ and $W$:
$$\mathrm{Id}_V: V \to V, \hspace{.5in} \mathrm{Id}_W: W \to W$$
These are linear transformations which simply output the input fed to them.

We can form the commutative squares associated with these maps:



We call the matrices like $\mathrm{Rep}_{\mathcal{B}\mathcal{B}'}(\mathrm{Id}_V)$ which represent the identity linear transformation with respsect to a pair of bases, a **change of basis** matrix.

Fix a vector $\vec{v} \in V$.  Then we can compute the representation of $\vec{v}$ with respect to the basis $\mathcal{B}$ to obtain the column vector:
$$\mathrm{Rep}_{\mathcal{B}}(\vec{v}) \in \mathbb{R}^n$$
If we wish to, instead, obtain the representation of $\vec{v}$ with respect to some other basis $\mathcal{B}'$, then we can find it by multiplying $\mathrm{Rep}_{\mathcal{B}}(\vec{v})$ by the change of basis matrix $\mathrm{Rep}_{\mathcal{B}\mathcal{B}'}(\mathrm{Id}_V)$ on the left.  This results 
$$\mathrm{Rep}_{\mathcal{B}\mathcal{B}'}(\mathrm{Id}_V) \cdot \mathrm{Rep}_{\mathcal{B}}(\vec{v}) = \mathrm{Rep}_{\mathcal{B}'}(\vec{v})$$

We can put together all of the commutative diagrams for the various linear transformations into a single diagram:



- Notice how the inner square of this diagram is the commutative square for the representation of $L$ with respect to $\mathcal{B}$ and $\mathcal{D}$.
- Notice how the commutative square for the representation of $\mathrm{Id}_V$ with respect to $\mathcal{B}$ and $\mathcal{B'}$ has become a trapezoid and is on the left side of the diagram.  The arrow pointing southwest is the change of basis matrix from $\mathcal{B}$ to $\mathcal{B}'$. 
- Notice how the commutative square for the representation of $\mathrm{Id}_W$ with respect to $\mathcal{D}$ and $\mathcal{D'}$ has become a trapezoid and is on the left side of the diagram.  The arrow pointing southeast is the change of basis matrix from $\mathcal{D}$ to $\mathcal{D}'$. 
- Notice how the outer square of this diagram is the commutative square for the representation of $L$ with respect to $\mathcal{B}'$ and $\mathcal{D}'$.  

Parsing this diagram gives us a nice formula for $\mathrm{Rep}_{\mathcal{B}'\mathcal{D}'}(L)$ in terms of $\mathrm{Rep}_{\mathcal{B}\mathcal{D}}(L)$.  In particular, focus on the bottom trapezoid.  Notice that starting at $\mathbb{R}^n$ in the lower left corner, we can traverse the arrow that is pointing "southwest" (though we travel northeast along it) to arrive at the $\mathbb{R}^n$ in the lower left corner of the inner square by doing the inverse of the change of basis matrix, that is $\mathrm{Rep}_{\mathcal{B}\mathcal{B}'}(\mathrm{Id}_V)^{-1}$. Then we can go from $\mathbb{R}^n$ to $\mathbb{R}^m$ by the map $\mathrm{Rep}_{\mathcal{B}\mathcal{D}}(L)$, and then arrive at the $\mathbb{R}^m$ in the lower right of the diagram by using the change of basis matrix $\mathrm{Rep}_{\mathcal{D}\mathcal{D}'}(\mathrm{Id}_W)$.  The commutativity of the bottom trapezoid tells us that:
$$\mathrm{Rep}_{\mathcal{B}'\mathcal{D}'}(L) = \mathrm{Rep}_{\mathcal{D}\mathcal{D}'}(\mathrm{Id}_W) \cdot \mathrm{Rep}_{\mathcal{B}\mathcal{D}}(L) \cdot \mathrm{Rep}_{\mathcal{B}\mathcal{B}'}(\mathrm{Id}_V)^{-1} $$

This notation is a bit out of hand, so we frequently shorten the expressions by denoting various matrices as:
$$P = \mathrm{Rep}_{\mathcal{B}\mathcal{B}'}(\mathrm{Id}_V)$$
$$Q = \mathrm{Rep}_{\mathcal{D}\mathcal{D}'}(\mathrm{Id}_W)$$
$$A = \mathrm{Rep}_{\mathcal{B}\mathcal{D}}(L)$$
$$A' = \mathrm{Rep}_{\mathcal{B}'\mathcal{D}'}(L)$$
Then the above result becomes:
$$\boxed{A' = Q \cdot A \cdot P^{-1}}$$

One important special case is when $A$ is a square matrix, $\mathcal{B} = \mathcal{D}$ and $\mathcal{B}' = \mathcal{D}'$. Of course, this implies that $P = Q$, so the above formula becomes:
$$\boxed{A' = P \cdot A \cdot P^{-1}}$$
where $P$ is the change of basis matrix from $\mathcal{B}$ to $\mathcal{B}'$.

## Eigenvectors and Eigenvalues

Given a square matrix $A$, an **eigenvector with eigenvalue $\lambda$** is a non-zero vector $\vec{v}$ with the property that:
$$A \vec{v} = \lambda \vec{v}$$

### Do eigenvectors always exist?

This is a bit of a delicate question that requires quite a bit of careful discussion. For the purposes of this notebook, just take the following as fact whose justification will come later in the course: **almost every square matrix has a basis of eigenvectors, though sometimes the eigenvalues are complex numbers.** 

This is sort of similar to the situation with non-singular matrices -- if you choose a matrix "at random," whatever that means, then you will get one with a "full set" of eigenvectors (by this I mean a basis of eigenvectors).  But don't be fooled!  Whether a matrix is non-singular or not is unrelated to whether it has a full set of eigenvectors.

### How to compute eigenvectors using numpy

Given a square numpy array `A`, we can access eigenvectors and eigenvalues by using the numpy command `np.linalg.eig(A)`.  Note that this command returns *two* arrays, arranged as a list (technically, a namedtuple -- but we can treat this data type as a list):
- `np.linalg.eig(A)[0]` is a list (technically a one-dimensional numpy array) of the eigenvalues of $A$.
- `np.linalg.eig(A)[1]` is a matrix whose columns are eigenvectors of $A$.  These eigenvectors are normalized so that the sum of the squares of the entries are 1 (note, you can multiply any eigenvector by a non-zero scalar and again obtain an eigenvector).
- `np.linalg.eig(A)[1][:,[i]]` is the $i$th column of A.  Its eigenvalue is `np.linalg.eig(A)[0][i]`.

## The Spectral Theorem

**Theorem:** If $A$ is a $n\times n$ symmetric matrix, then $\mathbb{R}^n$ with the standard inner product has an orthonormal basis consisting of eigenvectors of $A$.  

**Corollary:** If $A$ is a symmetric matrix, then there is an invertible matrix $P$ such that:
$$P A P^{T} = D$$
where $D$ is a diagonal matrix.  The entries along the diagonal of $D$ are all real numbers, and they are precisely the eigenvalues of $A$.
